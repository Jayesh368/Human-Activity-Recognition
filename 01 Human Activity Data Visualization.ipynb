```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import os
```

```python
data = pd.read_csv('Processed_DataSet/OriginalProcessedData.csv')

data.plot(subplots = True, sharex = True, layout = (30, 4) , figsize=(20,15),legend = False)
plt.show()
```

```python
data[data['Activity']=='Walking'].plot(legend=False)
```

```python
def VisualizeData1(Directory):
    dirs = os.listdir(Directory)
    for i in dirs:
        data=pd.read_csv(Directory + '\\' + i)
        data.plot(subplots = True, sharex = True, layout = (10,3) ,title=i.split('.')[0],figsize=(15,10) )
        
```

```python
VisualizeData1('OriginalRawDataset')
```

```python
VisualizeData1('NewRawDataset')
```
